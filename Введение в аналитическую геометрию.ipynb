```python
#окружность

a = 2.
b = 2.
axes()
plt.contour(x, y,(x**2/a**2 + y**2/b**2), [1], colors='k')
plt.show()
```

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D

fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 2)
Y = np.arange(-50, 50, 20)
print(X)
X, Y = np.meshgrid(X, Y)
print(Y)
print(X[1,1])
Z = 1 - X*2 - Y*3
Z2 = -12 - 2*X - 3*Y
ax.plot_surface(X, Y, Z)
ax.plot_surface(X, Y, Z2)
ax.scatter(0, 0, 0,'z',50,'red')
show()
```

```python
import numpy as np

ax = axes3d.Axes3D(plt.figure())

i = np.arange(-1, 1, 0.01)
X, Y = np.meshgrid(i, i)
Z = X**2 - Y**2

ax.plot_wireframe(X, Y, Z, rstride=10, cstride=10)

plt.show()
```

```python
ax = axes3d.Axes3D(plt.figure())

i = np.arange(-1, 1, 0.01)
X, Y = np.meshgrid(i, i)
Z = X**2 + Y**2

ax.plot_wireframe(X, Y, Z, rstride=10, cstride=10)

plt.show()
```
